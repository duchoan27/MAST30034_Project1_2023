# Data Preprocessing

### Weather data
First we preprocess the weather dataframe. 

We separate the time into 2 columns represent date and time. We also remove the extra column.

```python
times = weather_2023["time"].apply(lambda x: x[-5:-3])
dates = weather_2023["time"].apply(lambda x: x[:-6])
```

```python
times
```

```python
dates
```

```python
weather_2023['time'] = times
weather_2023['date'] = dates
```

The processed weather dataframe:

```python
weather_2023
```

### Yellow Cab Data

First we join data from three months together.

```python
yellow_df = pd.concat([yellow_aug_df, yellow_sep_df, yellow_oct_df])
yellow_df.reset_index(inplace=True)
```

```python
yellow_df.drop(columns=['index'], inplace=True)
```

```python
yellow_df
```

#### NaN values

Inspect the NaN value in the yellow cab dataframe.

```python
nan_count = yellow_df.isna().sum()/yellow_df.shape[0] *100
nan_count
```

Since the NaN values are not in the columns that we are considering for our model and the missing percentage is less than 5%, we do not remove or fill these cells.

#### Unreasonable numerical values

Check for any unreasonable numerical values.

```python
cols = ["passenger_count", "trip_distance", "fare_amount", "tip_amount", "tolls_amount", "total_amount"]
yellow_df[cols][yellow_df[cols] < 0].count()
```

**Value less than 0**

We can see that there is some value in the columns *fare_amount, tip_amount, tolls_amount, total_amount* is less than 0. This is unreasonable so we remove the instances that have those values less than 0.

```python
yellow_df[cols] = yellow_df[yellow_df[cols] >= 0][cols]
yellow_df.dropna(inplace=True)
```

Check if the all of the unreasonable value are removed.

```python
cols = ["passenger_count", "trip_distance", "fare_amount", "tip_amount", "tolls_amount", "total_amount"]
yellow_df[cols][yellow_df[cols] < 0].count()
```

**Passenger Count with 0 value**

We will remove instances with trip number of passenger equal to .

```python
yellow_df["passenger_count"] = yellow_df[yellow_df["passenger_count"] > 0]["passenger_count"]
yellow_df.dropna(inplace=True)
```

```python
yellow_df[yellow_df["passenger_count"] == 0].count()["passenger_count"]
```

**Trip Distance with 0 value**

We will remove instances with trip distance equal to 0.

```python
yellow_df["trip_distance"] = yellow_df[yellow_df["trip_distance"] > 0]["trip_distance"]
yellow_df.dropna(inplace=True)
```

```python
yellow_df[yellow_df["trip_distance"] == 0].count()["trip_distance"]
```

**Total Amount with 0 value**

We will remove intances with the total amount of money that the passenger has to pay equal to 0.

```python
yellow_df["total_amount"] = yellow_df[yellow_df["total_amount"] > 0]["total_amount"]
yellow_df.dropna(inplace=True)
```

```python
yellow_df[yellow_df["total_amount"] == 0].count()["total_amount"]
```

**Remove Location ID with value outside of our given region**

There are only 263 given regions in the dataset. So we remove instances with pick-up or drop-off locationID with value larger than 263.

```python
yellow_df["PULocationID"] = yellow_df[yellow_df["PULocationID"] <= 263]["PULocationID"]
yellow_df.dropna(inplace=True)
```

```python
yellow_df[yellow_df["PULocationID"] > 263].count()["PULocationID"]
```

```python
yellow_df["DOLocationID"] = yellow_df[yellow_df["DOLocationID"] <= 263]["DOLocationID"]
yellow_df.dropna(inplace=True)
```

```python
yellow_df[yellow_df["DOLocationID"] > 263].count()["DOLocationID"]
```

#### Related features

Since we care about the demand for yellow cabs in given regions based on time, we only select pick-up time and pick-up location

```python
yellow_final_df = yellow_df[["tpep_pickup_datetime", "PULocationID"]].copy()
yellow_final_df
```

Separate date and time data and convert pick-up location to integer value

```python
yellow_final_df["time"] = yellow_final_df["tpep_pickup_datetime"].apply(lambda x: str(x)[-8:-6])
yellow_final_df["date"] = yellow_final_df["tpep_pickup_datetime"].apply(lambda x: str(x)[:-9])
yellow_final_df["PULocationID"] = yellow_final_df["PULocationID"].apply(lambda x: int(x))
yellow_final_df
```

```python
yellow_final_df.drop(columns=['tpep_pickup_datetime'], inplace=True)
yellow_final_df
```

**Merge with the weather data**

```python
weather_2023
```

```python
trips_df = pd.merge(yellow_final_df, weather_2023, on=['time','date'], how='left')
```

```python
trips_df.isna().sum()
```

```python
trips_df.dropna(inplace=True)
```

```python
trips_df
```
